Dataset Description

This dataset contains yield and population per subcounty.  The dataset for the above questions can be found here. [Link]. 

The glossary for this table is as follows:

POP: total population for the subcounty
S_Yield_Ha: average yield for sorghum for the subcounty (Kg/Ha)
M_Yield_Ha: average yield for maize for the subcounty (Kg/Ha)
Crop_Area_Ha: total crop area for the subcounty (Ha)
S_Area_Ha: total sorghum crop area for the subcounty (Ha)
M_Area_Ha: total maize crop area for the subcounty (Ha)
S_Prod_Tot: total productivity for the sorghum for the subcounty (Kg)
M_Prod_Tot: total productivity for the maize for the subcounty (Kg)

```python
# Importing the python libraries needed in this notebook

import csv
import pandas as pd
```

```python
#Loading the Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv table from  local CSV file
#Colaboratory by clicking View > Table of Contents > Files then Upload. (NB: Do not load file from url)

with open('Uganda_Karamoja_Subcounty_Crop_Yield_Population.csv','r') as f:
    Uganda_Karamoja_Subcounty_Crop_Yield_Population = pd.read_csv(f, index_col=0, encoding='utf-8') 

# Then storing it in an SQL table of our in memory sqlite database

%sql DROP TABLE EXISTS Uganda_Karamoja_Subcounty_Crop_Yield_Population;
%sql PERSIST Uganda_Karamoja_Subcounty_Crop_Yield_Population;
```

```python
# To preview attributes of the dataset
%sql select * from Uganda_Karamoja_Subcounty_Crop_Yield_Population limit 10;
```

```python
# To Display a list of Sub Counties and their population and areas
%%sql
SELECT SUBCOUNTY_NAME, POP, Crop_Area_Ha FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population; 
```

```python
# To Sort the list of districts by total crop area (descending order)
%%sql
SELECT DISTRICT_NAME, Crop_Area_Ha FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
ORDER BY Crop_Area_Ha DESC;

```

```python
#To Select only the Sub counties from the Moroto district, order them alphabetically and show their production of sorghum
%%sql
SELECT SUBCOUNTY_NAME, S_Prod_tot, DISTRICT_NAME FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE DISTRICT_NAME = "MOROTO" 
ORDER BY SUBCOUNTY_NAME ASC;



```

```python
#Compute the total Maize production per District
%%sql
SELECT DISTRICT_NAME,SUM(M_Prod_Tot)
FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
GROUP BY DISTRICT_NAME

```

```python
#Compute the number of Sub counties where Maize is produced and the total Maize production per District
%%sql
SELECT DISTRICT_NAME, SUM(M_Prod_Tot), COUNT(SUBCOUNTY_NAME)
FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE "M_Prod_Tot" IS NOT NULL
GROUP BY "DISTRICT_NAME"
ORDER BY COUNT(SUBCOUNTY_NAME);
```

```python
#Compute the overall Crop area in all Sub counties where population is over 20000.
%%sql
SELECT SUM(Crop_Area_Ha) ,COUNT(SUBCOUNTY_NAME)
FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE POP > 20000;


```

```python
#Sort the Maize production in descending order by Districts, only taking into account Sub counties where Maize area is larger than Sorghum area, and display the number of Sub counties per district matching that criteria
#display the number of Sub counties per district matching that criteria
%%sql
SELECT DISTRICT_NAME,COUNT(SUBCOUNTY_NAME), M_Area_Ha,S_Area_Ha
FROM Uganda_Karamoja_Subcounty_Crop_Yield_Population
WHERE M_Area_Ha > S_Area_Ha 
GROUP BY DISTRICT_NAME
ORDER BY M_Prod_Tot DESC;






```
